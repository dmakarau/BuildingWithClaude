```python
# Import the reusable helper module
from claude_helpers import client, model, add_user_message, add_assistant_message, chat
```

```python
# Initial list of messages
messages = []

while True:
    # Get user input
    user_input = input("> ")
    print("> ", user_input)
    add_user_message(messages, user_input)
    # Get assistant response
    assistant_response = chat(messages)
    # Add assistant response to messages and print it
    add_assistant_message(messages, assistant_response)
    print("-------------")
    print("> ", assistant_response)
    print("-------------")
```
