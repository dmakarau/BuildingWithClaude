```python
# Load environment variable
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client 
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-5-20250929"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model = model,
        max_tokens = 1000,
        messages = messages,
    )
    return message.content[0].text
```

```python
# Make a starting list of messages
messages = []

# Add initial user question "Define quantum computing in one sentence."
add_user_message(messages, "Define quantum computing in one sentence.")

# Pass the list of messages to the chat function to get the answer
answer = chat(messages)

# Take the answer and add it to the list of messages as an assistant message
add_assistant_message(messages, answer)

# Add a follow-up user question
add_user_message(messages, "Write another sentence")

# Pass the updated list of messages to the chat function to get the follow-up answer
follow_up_answer = chat(messages)
follow_up_answer
```
